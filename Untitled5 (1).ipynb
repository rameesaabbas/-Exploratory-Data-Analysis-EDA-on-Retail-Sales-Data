TIME SERIES

```python
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)
```

```python
monthly_sales = df['Total Amount'].resample('M').sum()
```

TREND OVER TIME

```python
plt.figure(figsize=(10, 6))
plt.plot(monthly_sales, marker='o', linestyle='-')
plt.title('Monthly Sales Trends')
plt.xlabel('Date')
plt.ylabel('Total Sales Amount')
plt.grid(True)
plt.show()
```

```python
top_products = df.groupby('Product Category')['Total Amount'].sum().sort_values(ascending=False).head(5)
print("Top Selling Products:")
print(top_products)
```

```python
if 'Date' in df.columns:
    df['YearMonth'] = df['Date'].dt.to_period('M')
    monthly_sales = df.groupby('YearMonth')['Total Amount'].sum()
    plt.figure(figsize=(10, 6))
    plt.plot(monthly_sales, marker='o', linestyle='-')
    plt.title('Monthly Sales Trends')
    plt.xlabel('Year-Month')
    plt.ylabel('Total Sales Amount')
    plt.grid(True)
    plt.xticks(rotation=45)
    plt.show()
```

VISUALIZATION

```python
data = {
    'Year': [2018, 2019, 2020, 2021, 2022],
    'Sales': [1000, 1500, 2000, 1800, 2200],
    'Expenses': [800, 1000, 1200, 1100, 1300],
    'Profit': [200, 500, 800, 700, 900]
}
```

```python
df = pd.DataFrame(data)
```

BAR CHART

```python
plt.figure(figsize=(10, 6))
plt.bar(df['Year'], df['Sales'], color='blue', label='Sales')
plt.bar(df['Year'], df['Expenses'], color='red', label='Expenses')
plt.xlabel('Year')
plt.ylabel('Amount ($)')
plt.title('Sales and Expenses Over Years')
plt.legend()
plt.show()
```

LINE PLOT

```python
plt.figure(figsize=(10, 6))
plt.plot(df['Year'], df['Sales'], marker='o', linestyle='-', color='green', label='Sales')
plt.plot(df['Year'], df['Expenses'], marker='o', linestyle='-', color='red', label='Expenses')
plt.xlabel('Year')
plt.ylabel('Amount ($)')
plt.title('Sales and Expenses Over Years')
plt.legend()
plt.grid(True)
plt.show()
```

HEATMAP

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df[['Sales', 'Expenses', 'Profit']], annot=True, cmap='coolwarm', fmt=".1f")
plt.title('Heatmap of Sales, Expenses, and Profit')
plt.xlabel('Metrics')
plt.ylabel('Year')
plt.show()
```
